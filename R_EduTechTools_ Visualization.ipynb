##
Over the last 1.5 years, education has undergone one of the greatest metamorphoses 
in its history. Even 50 years ago, it was unthinkable that students in schools would use computers
 for educational purposes, but what happened in 2020 showed only how necessary the computerization of 
society may be. The outbreak of the COVID-19  has forced billions of people to limit direct contact with each other. 
Many companies have switched to remote work, 
and commerce has largely moved to the Internet. Schools, 
universities and other educational institutions did the same - they switched to remote learning 
where teachers conduct lessons via the Internet for their students. 

It was certainly a challenge for teachers and students, but many ready-made tools that helped in 
conducting classes and lectures came to the rescue. 
Rapidly developing before the pandemic and gaining in popularity, e-learning has thus become the best form 
of education in these strange times. 
Always want to look at perspective of the specific tools used in the USA in 2020, 
thus checking: when, where, how and why specific learning tools were used. 
Who's the winner for 2020? Are those the tool our teacher use often? Cross US, are less less 

pct_black/hispanic - Percentage of students in the districts identified as Black or Hispanic based on 2018-19 NCES data
countyconnectionsratio - ratio (residential fixed high-speed connections over 200 kbps in at least one direction/households) based on the county level data from FCC From 477 (December 2018 version). See FCC data for more information.

1) What is the picture of digital connectivity and engagement in 2020?
2) What is the effect of the COVID-19 pandemic on online and distance learning, and how might this also evolve in the future?
3) How does student engagement with different types of education technology change over the course of the pandemic?
4) How does student engagement with online learning platforms relate to different geography? Demographic context (e.g., race/ethnicity, ESL, learning disability)? Learning context? Socioeconomic status?
5) Do certain state interventions, practices or policies (e.g., stimulus, reopening, eviction moratorium) correlate with the increase or decrease online engagement?

```r
products <- read.csv("../Python Projects/products_info.csv")
head(products)
names(products)[1] <- "lp_id"

districts <- read.csv("../Python Projects/districts_info.csv")
head(districts)

files <- list.files("../Python Projects/engagement_data")
files2 <- paste0("../Python Projects/engagement_data/", files)
engage <- lapply(files2, read_csv) %>%
  bind_rows(.id = "source")
head(engage)
```

```r
head(products)
tail（engage)
head(districts)
```

```r
disctrict_joiner <- data.frame(district_id = districts$district_id, source = as.character(1:nrow(districts)))
engage <- engage %>%
  left_join(disctrict_joiner, by = "source")

engage <- engage %>%
  left_join(districts, by = "district_id")

engage <- engage %>%
  left_join(products, by = "lp_id")

engage <- engage[complete.cases(engage$Product.Name),]

engage$engagement_index[is.na(engage$engagement_index)] <- 0

head(engage)
```

```r
engage %>%
  group_by(Product.Name) %>%
  summarise(mean = mean(engagement_index)/1000) %>%
  arrange(desc(mean)) %>%
  slice(1:50) %>%
ggplot(., aes(reorder(Product.Name, +mean), mean))+
  geom_bar(col = "gray10", stat = "identity", width = 0.75, fill = "limegreen", alpha = 1)+
  coord_flip()+
  labs(title = "Mean daily page-load events in top 40 tools", subtitle = "per 1 student", y = "Mean daily page-load events per one student", 
       x = "Product")+
  theme_fivethirtyeight()+
  theme_michau
set_plot_dimensions(22,25)
```

```r
#http://www.di.fc.ul.pt/~jpn/r/GraphicalTools/colorPalette.html#rcolorbrewer
engage %>%
  filter(Product.Name %in% c("Google Docs","Seesaw : The Learning Journal", "Google Classroom", "YouTube", "Canvas", "Meet","Schoology","Kahoot!")) %>%
  group_by(Product.Name,  time = week(as.Date(time))) %>%
  summarise(mean = mean(engagement_index)/1000, .groups = "drop") %>%
ggplot(., aes(time, mean, colour = Product.Name))+
  geom_line(size = 1.2, alpha = 1)+
  scale_colour_brewer(palette = "Dark2")+
  geom_vline(xintercept = 11, linetype = "longdash", size = 0.35, col = "gray30")+
  geom_vline(xintercept = 36, linetype = "longdash", size = 0.35, col = "gray30")+
  annotate("text", x = 6.5, y = 16.1, label = "1,000 cases \n of COVID", size = 5.4)+
  annotate("text", x = 31.5, y = 15.8, label = "1st September", size = 5.4)+
  annotate(geom = "curve", x = 10.7, y = 12.9, xend = 6.9, yend = 15, curvature = -0.3, arrow = arrow(length = unit(5, "mm")))+
  annotate(geom = "curve", x = 35.7, y = 12.9, xend = 31.9, yend = 15, curvature = -0.3, arrow = arrow(length = unit(5, "mm")))+
  labs(title = "Mean daily page-load events in top 7 tools", subtitle = "by products and time, per 1 student", y = "Mean daily page-load events per one student", 
       x = "Week of year (2020)", colour = "Product: ")+
  theme_fivethirtyeight()+
  theme_michau+
  theme(legend.position = "right", legend.direction = "vertical")
```

```r
engage %>%
  filter(pct_black.hispanic != "NA") %>%
  filter(Product.Name %in% c("Google Docs", "Lexia Core5 Reading","Google Classroom", "YouTube", "Canvas", "Meet", "Schoology", "Kahoot!", "Google Forms", "Google Drive", "ClassLink","Zoom","Clever","Seesaw : The Learning Journal")) %>%
  group_by(Product.Name, pct_black.hispanic) %>%
  summarise(mean = mean(engagement_index)/1000, .groups = "drop") %>%
  mutate(color = ifelse(mean > 5,"0", "1")) %>%
  arrange(pct_black.hispanic) %>%
ggplot(., aes(pct_black.hispanic, mean, fill = color))+
  geom_bar(col = "gray10", stat = "identity", width = 0.5, alpha = 0.8)+
  facet_wrap(.~Product.Name, nrow = 2)+
  scale_fill_brewer(palette = "Set4")+
  scale_y_continuous(breaks = c(0, 5, 10, 15))+
  coord_flip()+
  labs(title = "Mean daily page-load events in top 14 tools", subtitle = "by share of black/hispanic students and products, per 1 student", y = "Mean daily page-load events per one student", 
       x = "Share of black/hispanic students")+
  theme_fivethirtyeight()+
  theme_michau+
  theme(legend.position = "none")
set_plot_dimensions(6,10)
```

```r
engage %>%
  filter(county_connections_ratio != "NA") %>%
  filter(Product.Name %in% c("Google Docs", "Lexia Core5 Reading","Google Classroom", "YouTube", "Canvas", "Meet", "Schoology", "Kahoot!", "Google Forms", "Google Drive", "ClassLink","Zoom","Clever","Seesaw : The Learning Journal")) %>%
  group_by(Product.Name, county_connections_ratio) %>%
  summarise(mean = mean(engagement_index)/1000, .groups = "drop") %>%
  mutate(color = ifelse(mean > 5,"0", "1")) %>%
  arrange(county_connections_ratio) %>%
ggplot(., aes(county_connections_ratio, mean, fill = color))+
  geom_bar(col = "gray10", stat = "identity", width = 0.5, alpha = 0.8)+
  facet_wrap(.~Product.Name, nrow = 2)+
  scale_fill_brewer(palette = "Set4")+
  scale_y_continuous(breaks = c(0, 5, 10, 15))+
  coord_flip()+
  labs(title = "Mean daily page-load events in top 14 tools", subtitle = "by ratio of residential fixed high-speed connections/households and products, per 1 student", y = "Mean daily page-load events per one student", 
       x = "Ratio of residential fixed high-speed connections/households")+
  theme_fivethirtyeight()+
  theme_michau+
  theme(legend.position = "none")
set_plot_dimensions(15,10)
```

```r
engage %>%
  filter(locale != "NA") %>%
  filter(Product.Name %in% c("Google Docs", "Lexia Core5 Reading","Google Classroom", "YouTube", "Canvas", "Meet", "Schoology", "Kahoot!", "Google Forms", "Google Drive", "ClassLink","Zoom","Clever","Seesaw : The Learning Journal")) %>%
  group_by(Product.Name, locale ) %>%
  summarise(mean = mean(engagement_index)/1000, .groups = "drop") %>%
  mutate(color = ifelse(mean > 5,"0", "1")) %>%
  arrange(locale) %>%
ggplot(., aes(locale, mean, fill = color))+
  geom_bar(col = "gray10", stat = "identity", width = 0.75, alpha = 0.8)+
  facet_wrap(.~Product.Name, nrow = 2)+
  scale_fill_brewer(palette = "Set4")+
  scale_y_continuous(breaks = c(0, 5, 10, 15))+
  coord_flip()+
  labs(title = "Mean daily page-load events in top 14 tools", subtitle = "by ratio of residential fixed high-speed connections/households and products, per 1 student", y = "Mean daily page-load events per one student", 
       x = "locale")+
  theme_fivethirtyeight()+
  theme_michau+
  theme(legend.position = "none")
set_plot_dimensions(15,10)
```

```r
engage %>%
  filter(state %in% c( "Massachusetts", "Michigan", "Minnesota", "Missouri", "New Hampshire","California","District Of Columbia", "Florida", "Illinois")) %>%
  filter(Product.Name %in% c("Google Docs", "Lexia Core5 Reading","Google Classroom", "YouTube", "Canvas", "Meet", "Schoology", "Kahoot!", "Google Forms", "Google Drive", "ClassLink","Zoom","Clever","Seesaw : The Learning Journal")) %>%
  group_by(Product.Name, state) %>%
  summarise(mean = mean(engagement_index)/1000, .groups = "drop") %>%
  arrange(desc(mean)) %>%
  mutate(color = ifelse(mean > 5,"0", "1")) %>%
ggplot(., aes(reorder(Product.Name, +mean), mean, fill = color))+
  geom_bar(col = "gray10", stat = "identity", width = 0.5, alpha = 0.9)+
  facet_wrap(.~state)+
  scale_fill_brewer(palette = "Set6")+
  scale_y_continuous(breaks = c(0, 5, 10, 15))+
  coord_flip()+
  labs(title = "Mean daily page-load events in top 10 tools", subtitle = "by product and state (I to N), per 1 student", y = "Mean daily page-load events per one student", 
       x = "Product")+
  theme_fivethirtyeight()+
  theme_michau+
  theme(legend.position = "none")
set_plot_dimensions(20,25)
```
